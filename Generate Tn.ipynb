```python
def check_prime(num: int):
    if num > 1:

        for i in range(2, num):
            if (num % i) == 0:
                return False

    return True

for x in range(10):
    print(x, check_prime(x))
```

```python
def gen_tn(low=0, high=100, n=3, verbose=True):
    all_nums = []
    for j, i in enumerate(range(low, high, n)):
        c = bin(i).count('1') % 2
        if verbose:
            prime = check_prime(j)
            if prime:
                print(f"{prime=} {i}: {c}")
        all_nums.append(c)
    return all_nums
```

```python
import pandas as pd
df = pd.read_csv("./rl-locations-new/data4/data_locations_of_ones.csv")
df = df.head(10000)
for x in df["loc_in_t"]:
    print(x)
```

```python
def return_under(df):
    all = []
    total = len(df)
    for x in range(0, total):
        p = check_prime(df.iloc[x].loc_in_t)
        if p:
            all.append(True)
        else:
            all.append(False)

    return df[all]

df = return_under(df)
print(df.head())
```
